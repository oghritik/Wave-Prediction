```python
# ==============================================================================
# All-in-One Cell for a < 2 Minute Test Run
# ==============================================================================

# Step 0: Environment Setup and Library Imports
# ------------------------------------------------------------------------------
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tqdm.auto import tqdm
import warnings
import time

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
print(f"PyTorch Version: {torch.__version__}")

# Set up the GPU device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    torch.cuda.empty_cache()

# Step 1 & 2: Data Loading & Preprocessing
# ------------------------------------------------------------------------------
print("\n--- Step 1 & 2: Loading and Preprocessing Data ---")
data_path = r'D:\babe_prediction\cmems_mod_ibi_wav_anfc_0.027deg_PT1H-i_multi-vars_11.00W-8.53W_38.50N-40.97N_2022-12-01-2025-01-01-001.nc'
try:
    ds = xr.open_dataset(data_path, engine="netcdf4")
    print("Dataset loaded successfully!")
except FileNotFoundError:
    raise RuntimeError(f"Dataset not found. Please check the path: {data_path}")

# --- Feature Engineering & Data Preparation ---
target_var = 'VHM0'
initial_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR', 'VTPK', 'VSDX', 'VSDY']
final_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR_sin', 'VMDR_cos', 'VTPK', 'VSDX', 'VSDY']
all_vars_to_process = [target_var] + initial_feature_vars

# Convert to float32 to save memory and fill NaNs
ds_processed = ds.astype(np.float32).fillna(0)

# --- Data Scaling (fitting scalers on the full dataset) ---
scalers = {}
for var in all_vars_to_process:
    if var == 'VMDR': continue
    scaler = MinMaxScaler()
    data_2d = ds_processed[var].values.reshape(ds_processed[var].shape[0], -1)
    scaler.fit(data_2d)
    scalers[var] = scaler
print("Scalers fitted successfully on the dataset.")

# Step 3: PyTorch Optimized Dataset and DataLoaders
# ------------------------------------------------------------------------------
print("\n--- Step 3: Creating Optimized DataLoaders with a TINY Subset ---")
# --- FIX: Replace your old dataset class with this corrected one ---
class WaveDataset_Optimized(Dataset):
    def __init__(self, xr_dataset, feature_vars, target_var, indices, scalers, lookback=24):
        self.ds = xr_dataset
        self.feature_vars = feature_vars
        self.target_var = target_var
        self.indices = indices
        self.lookback = lookback
        self.scalers = scalers

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        # Get the start and end index for the full sequence slice
        seq_start_idx = self.indices[idx]
        target_idx = seq_start_idx + self.lookback

        # Lazily select the time slice from the xarray dataset in memory
        data_slice = self.ds.isel(time=slice(seq_start_idx, target_idx + 1))
        
        # --- FIX: Add .values to extract the numpy array ---
        radians = np.deg2rad(data_slice['VMDR'].values)
        
        # Now the rest of the code works correctly
        data_slice['VMDR_sin'] = (('time', 'latitude', 'longitude'), np.sin(radians).astype(np.float32))
        data_slice['VMDR_cos'] = (('time', 'latitude', 'longitude'), np.cos(radians).astype(np.float32))
        
        for var, scaler in self.scalers.items():
            if var == 'VMDR': continue
            # --- FIX: Add .values here as well ---
            var_data = data_slice[var].values
            var_data_2d = var_data.reshape(var_data.shape[0], -1)
            var_scaled_2d = scaler.transform(var_data_2d)
            data_slice[var] = (data_slice[var].dims, var_scaled_2d.reshape(var_data.shape))

        # --- Stack features and get target ---
        features_list = [data_slice[var].isel(time=slice(0, self.lookback)).values for var in self.feature_vars]
        X = np.stack(features_list, axis=0).astype(np.float32)
        y = data_slice[self.target_var].isel(time=self.lookback).values.astype(np.float32)

        # Transpose to (lookback, features, lat, lon) and add channel dim to target
        X = X.transpose(1, 0, 2, 3) 
        y = np.expand_dims(y, axis=0)
        
        return torch.from_numpy(X), torch.from_numpy(y)

# --- Data Splitting (Using a very small number of samples for the test) ---
lookback_hours = 24
batch_size = 8
n_time = ds.dims['time']
n_sequences = n_time - lookback_hours
all_indices = np.arange(n_sequences)
train_end = int(n_sequences * 0.7)

# Using a very small, fixed number of samples for a < 2 min run
num_train_samples = 200
num_val_samples = 50
train_indices = all_indices[:num_train_samples]
val_indices = all_indices[train_end:train_end + num_val_samples]

print(f"--- USING A VERY SMALL SUBSET FOR FAST EXECUTION ---")
print(f"Using {len(train_indices)} training samples and {len(val_indices)} validation samples.")

# --- Instantiate DataLoaders with the FIX ---
train_dataset = WaveDataset_Optimized(ds_processed, final_feature_vars, target_var, train_indices, scalers, lookback_hours)
val_dataset = WaveDataset_Optimized(ds_processed, final_feature_vars, target_var, val_indices, scalers, lookback_hours)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0) # FIX: num_workers=0
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0) # FIX: num_workers=0

# Step 4: Build the ConvLSTM Model
# ------------------------------------------------------------------------------
print("\n--- Step 4: Building the ConvLSTM Model ---")
class ConvLSTMCell(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, bias):
        super(ConvLSTMCell, self).__init__()
        self.input_dim, self.hidden_dim, self.kernel_size, self.padding, self.bias = input_dim, hidden_dim, kernel_size, kernel_size[0] // 2, bias
        self.conv = nn.Conv2d(in_channels=self.input_dim + self.hidden_dim, out_channels=4 * self.hidden_dim, kernel_size=self.kernel_size, padding=self.padding, bias=self.bias)
    def forward(self, input_tensor, cur_state):
        h_cur, c_cur = cur_state; combined = torch.cat([input_tensor, h_cur], dim=1); combined_conv = self.conv(combined)
        cc_i, cc_f, cc_o, cc_g = torch.split(combined_conv, self.hidden_dim, dim=1)
        i, f, o, g = torch.sigmoid(cc_i), torch.sigmoid(cc_f), torch.sigmoid(cc_o), torch.tanh(cc_g)
        c_next = f * c_cur + i * g; h_next = o * torch.tanh(c_next)
        return h_next, c_next
    def init_hidden(self, batch_size, image_size, device):
        height, width = image_size
        return (torch.zeros(batch_size, self.hidden_dim, height, width, device=device), torch.zeros(batch_size, self.hidden_dim, height, width, device=device))

class ConvLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, num_layers, batch_first=False, bias=True):
        super(ConvLSTM, self).__init__()
        self.batch_first = batch_first; self.num_layers = num_layers; hidden_dim = [hidden_dim] * num_layers if isinstance(hidden_dim, int) else hidden_dim
        cell_list = [ConvLSTMCell(input_dim=input_dim if i == 0 else hidden_dim[i - 1], hidden_dim=hidden_dim[i], kernel_size=kernel_size, bias=bias) for i in range(num_layers)]
        self.cell_list = nn.ModuleList(cell_list)
    def forward(self, input_tensor, hidden_state=None):
        b, _, _, h, w = input_tensor.size(); hidden_state = self._init_hidden(b, (h, w), input_tensor.device) if hidden_state is None else hidden_state
        layer_output_list = []; last_state_list = []
        seq_len = input_tensor.size(1); cur_layer_input = input_tensor
        for layer_idx in range(self.num_layers):
            h, c = hidden_state[layer_idx]; output_inner = []
            for t in range(seq_len):
                h, c = self.cell_list[layer_idx](input_tensor=cur_layer_input[:, t, :, :, :], cur_state=[h, c])
                output_inner.append(h)
            layer_output = torch.stack(output_inner, dim=1); cur_layer_input = layer_output
            layer_output_list.append(layer_output); last_state_list.append([h, c])
        return layer_output_list[-1], last_state_list[-1]
    def _init_hidden(self, batch_size, image_size, device):
        return [cell.init_hidden(batch_size, image_size, device) for cell in self.cell_list]

class ConvLSTMNet(nn.Module):
    def __init__(self, input_dim, hidden_dims=[64, 32], kernel_size=(3, 3)):
        super(ConvLSTMNet, self).__init__()
        self.cl1 = ConvLSTM(input_dim, hidden_dims[0], kernel_size, 1, batch_first=True)
        self.cl2 = ConvLSTM(hidden_dims[0], hidden_dims[1], kernel_size, 1, batch_first=True)
        self.output_conv = nn.Conv2d(hidden_dims[1], 1, kernel_size=(1, 1), padding='same')
    def forward(self, x):
        l1_out, _ = self.cl1(x); l2_out, _ = self.cl2(l1_out)
        return self.output_conv(l2_out[:, -1, :, :, :])

model = ConvLSTMNet(len(final_feature_vars))
if torch.cuda.device_count() > 1: model = nn.DataParallel(model)
model.to(device)
print("Model built and enabled for multi-GPU.")

# Step 5: Train and Visualize
# ------------------------------------------------------------------------------
def weighted_mse_loss(output, target, weight_threshold=0.8):
    mse = (output - target)**2; weights = torch.where(target > weight_threshold, 2.0, 1.0)
    return torch.mean(weights * mse)

def visualize_epoch_prediction(model, val_loader, device, scalers, epoch):
    model.eval()
    with torch.no_grad():
        X_val, y_val = next(iter(val_loader)); X_val = X_val.to(device)
        pred_val = model(X_val)[0].cpu().numpy().squeeze()
        true_val = y_val[0].cpu().numpy().squeeze()
        ts = scalers[target_var]
        pred_orig = ts.inverse_transform(pred_val.reshape(1, -1)).reshape(pred_val.shape)
        true_orig = ts.inverse_transform(true_val.reshape(1, -1)).reshape(true_val.shape)
        
        fig, axes = plt.subplots(1, 3, figsize=(20, 5)); fig.suptitle(f'Epoch {epoch+1} - Sample Validation Prediction', fontsize=16)
        vmax = max(true_orig.max(), pred_orig.max())
        im1 = axes[0].imshow(true_orig, cmap='viridis', vmin=0, vmax=vmax); axes[0].set_title("Ground Truth"); plt.colorbar(im1, ax=axes[0], fraction=0.046, pad=0.04)
        im2 = axes[1].imshow(pred_orig, cmap='viridis', vmin=0, vmax=vmax); axes[1].set_title("Model Prediction"); plt.colorbar(im2, ax=axes[1], fraction=0.046, pad=0.04)
        err = pred_orig - true_orig; vmax_err = np.abs(err).max()
        im3 = axes[2].imshow(err, cmap='RdBu_r', vmin=-vmax_err, vmax=vmax_err); axes[2].set_title("Prediction Error"); plt.colorbar(im3, ax=axes[2], fraction=0.046, pad=0.04)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

def train_model(model, train_loader, val_loader, device, scalers, epochs=2, model_path='best_convlstm_model.pth'):
    optimizer = optim.Adam(model.parameters(), lr=1e-4); loss_fn = weighted_mse_loss
    best_val_loss = float('inf'); scaler = torch.cuda.amp.GradScaler()
    print("\n--- Step 5: Starting Model Training ---")
    for epoch in range(epochs):
        start_time = time.time(); model.train(); total_train_loss = 0.0
        train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Training]")
        for X, y in train_pbar:
            X, y = X.to(device), y.to(device)
            with torch.cuda.amp.autocast():
                output = model(X)
                loss = loss_fn(output, y)
            optimizer.zero_grad(); scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
            total_train_loss += loss.item()
            train_pbar.set_postfix({'loss': f'{loss.item():.6f}'})
        
        model.eval(); total_val_loss = 0.0
        with torch.no_grad():
            for X_v, y_v in val_loader:
                X_v, y_v = X_v.to(device), y_v.to(device)
                val_output = model(X_v)
                total_val_loss += loss_fn(val_output, y_v).item()
        
        avg_train_loss, avg_val_loss = total_train_loss / len(train_loader), total_val_loss / len(val_loader)
        print(f"Epoch {epoch+1} Summary | Avg Train Loss: {avg_train_loss:.6f} | Avg Val Loss: {avg_val_loss:.6f} | Time: {time.time() - start_time:.2f}s")
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            # Save model correctly with DataParallel
            state_dict = model.module.state_dict() if isinstance(model, nn.DataParallel) else model.state_dict()
            torch.save(state_dict, model_path)
            print(f"  -> âœ… New best model saved (Val Loss: {best_val_loss:.6f})")
        
        visualize_epoch_prediction(model, val_loader, device, scalers, epoch)
    print("\n--- Training finished! ---")

# --- Start the Training for 2 epochs for a very quick run ---
train_model(model, train_loader, val_loader, device, scalers, epochs=2)
```

```python
# ==============================================================================
# FINAL All-in-One Cell with Memory-Efficient Lazy Loading
# ==============================================================================

# Step 0: Environment Setup and Library Imports
# ------------------------------------------------------------------------------
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg') # Use non-GUI backend
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tqdm.auto import tqdm
import warnings
import time

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')
print(f"PyTorch Version: {torch.__version__}")

# Set up the GPU device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
if torch.cuda.is_available():
    torch.cuda.empty_cache()

# Step 1 & 2: Data Loading & Preprocessing
# ------------------------------------------------------------------------------
print("\n--- Step 1 & 2: Loading Data Path and Fitting Scalers ---")
data_path = r'D:\\babe_prediction\\cmems_mod_ibi_wav_anfc_0.027deg_PT1H-i_multi-vars_11.00W-8.53W_38.50N-40.97N_2022-12-01-2025-01-01-001.nc'
try:
    ds = xr.open_dataset(data_path, engine="netcdf4")
    print("Dataset metadata loaded successfully!")
except FileNotFoundError:
    raise RuntimeError(f"Dataset not found. Please check the path: {data_path}")

# --- Feature Engineering & Data Preparation ---
target_var = 'VHM0'
initial_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR', 'VTPK', 'VSDX', 'VSDY']
final_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR_sin', 'VMDR_cos', 'VTPK', 'VSDX', 'VSDY']
all_vars_to_process = [target_var] + initial_feature_vars

# --- Data Scaling (fitting scalers only, without loading all data to RAM) ---
# This part is memory-intensive but necessary. We'll release the memory right after.
temp_ds_for_scaling = ds.fillna(0)
scalers = {}
for var in all_vars_to_process:
    if var == 'VMDR': continue
    scaler = MinMaxScaler()
    scaler.fit(temp_ds_for_scaling[var].values.reshape(-1, 1))
    scalers[var] = scaler
del temp_ds_for_scaling # Release the memory
print("Scalers fitted successfully on the dataset.")


# Step 3: PyTorch Lazy Dataset and DataLoaders
# ------------------------------------------------------------------------------
print("\n--- Step 3: Creating DataLoaders with Custom Date Split (Lazy Loading) ---")
# UPDATED: Using the memory-efficient Lazy Dataset
class WaveDataset_Lazy(Dataset):
    def __init__(self, data_path, feature_vars, target_var, indices, scalers, lookback=24):
        self.data_path, self.feature_vars, self.target_var, self.indices, self.lookback, self.scalers = data_path, feature_vars, target_var, indices, lookback, scalers
        self.all_vars_on_disk = [target_var] + ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR', 'VTPK', 'VSDX', 'VSDY']

    def __len__(self): return len(self.indices)
    def __getitem__(self, idx):
        seq_start_idx = self.indices[idx]; target_idx = seq_start_idx + self.lookback
        with xr.open_dataset(self.data_path) as ds_slice:
            data_slice = ds_slice[self.all_vars_on_disk].isel(time=slice(seq_start_idx, target_idx + 1))
        data_slice = data_slice.fillna(0); radians = np.deg2rad(data_slice['VMDR'])
        data_slice['VMDR_sin'] = (('time', 'latitude', 'longitude'), np.sin(radians).astype(np.float32))
        data_slice['VMDR_cos'] = (('time', 'latitude', 'longitude'), np.cos(radians).astype(np.float32))
        for var, scaler in self.scalers.items():
            var_data = data_slice[var].values; var_data_2d = var_data.reshape(var_data.shape[0], -1)
            var_scaled_2d = scaler.transform(var_data_2d)
            data_slice[var] = (data_slice[var].dims, var_scaled_2d.reshape(var_data.shape))
        features_list = [data_slice[var].isel(time=slice(0, self.lookback)).values for var in self.feature_vars]
        X = np.stack(features_list, axis=0).astype(np.float32)
        y = data_slice[self.target_var].isel(time=self.lookback).values.astype(np.float32)
        X, y = X.transpose(1, 0, 2, 3), np.expand_dims(y, axis=0)
        return torch.from_numpy(X), torch.from_numpy(y)

# --- Data Splitting by Year ---
lookback_hours = 24
batch_size = 8
n_sequences = ds.dims['time'] - lookback_hours
all_start_indices = np.arange(n_sequences)
target_times = pd.to_datetime(ds.time.values[lookback_hours:])
val_indices = all_start_indices[(target_times.year >= 2020) & (target_times.year <= 2022)]
train_indices = all_start_indices[(target_times.year >= 2023) & (target_times.year <= 2025)]
print(f"Found {len(train_indices)} training samples (target years 2023-2025).")
print(f"Found {len(val_indices)} validation samples (target years 2020-2022).")

# --- Instantiate DataLoaders ---
train_dataset = WaveDataset_Lazy(data_path, final_feature_vars, target_var, train_indices, scalers, lookback_hours)
val_dataset = WaveDataset_Lazy(data_path, final_feature_vars, target_var, val_indices, scalers, lookback_hours)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=0)

# Step 4: Build the ConvLSTM Model
# ------------------------------------------------------------------------------
print("\n--- Step 4: Building the ConvLSTM Model ---")
class ConvLSTMCell(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, bias):
        super(ConvLSTMCell, self).__init__(); self.input_dim,self.hidden_dim,self.kernel_size,self.padding,self.bias = input_dim,hidden_dim,kernel_size,kernel_size[0]//2,bias
        self.conv = nn.Conv2d(self.input_dim+self.hidden_dim, 4*self.hidden_dim, self.kernel_size, padding=self.padding, bias=self.bias)
    def forward(self, x, h_c):
        h,c=h_c; cc=self.conv(torch.cat([x,h],dim=1)); i,f,o,g=torch.split(cc,self.hidden_dim,dim=1)
        i,f,o,g=torch.sigmoid(i),torch.sigmoid(f),torch.sigmoid(o),torch.tanh(g); c_n=f*c+i*g
        return o*torch.tanh(c_n), c_n
    def init_hidden(self, b, i_s, d): return (torch.zeros(b,self.hidden_dim,*i_s,device=d), torch.zeros(b,self.hidden_dim,*i_s,device=d))

class ConvLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, num_layers, batch_first=False, bias=True):
        super(ConvLSTM, self).__init__(); self.batch_first=batch_first; self.num_layers=num_layers
        h_d=[hidden_dim]*num_layers if isinstance(hidden_dim,int) else hidden_dim
        cl=[ConvLSTMCell(input_dim if i==0 else h_d[i-1],h_d[i],kernel_size,bias) for i in range(num_layers)]
        self.cell_list=nn.ModuleList(cl)
    def forward(self, x, h_c=None):
        b,_,_,h,w=x.size(); h_c=self._init_hidden(b,(h,w),x.device) if h_c is None else h_c
        out_l,last_s=[],[]; s_l=x.size(1); cur_in=x
        for l_idx in range(self.num_layers):
            h,c=h_c[l_idx]; out_i=[]
            for t in range(s_l):
                h,c=self.cell_list[l_idx](cur_in[:,t,:,:,:], [h,c]); out_i.append(h)
            l_out=torch.stack(out_i,dim=1); cur_in=l_out; out_l.append(l_out); last_s.append([h,c])
        return out_l[-1], last_s[-1]
    def _init_hidden(self, b, i_s, d): return [c.init_hidden(b,i_s,d) for c in self.cell_list]

class ConvLSTMNet(nn.Module):
    def __init__(self, input_dim, hidden_dims=[64, 32], kernel_size=(3, 3)):
        super(ConvLSTMNet, self).__init__(); self.cl1=ConvLSTM(input_dim,hidden_dims[0],kernel_size,1,True)
        self.cl2=ConvLSTM(hidden_dims[0],hidden_dims[1],kernel_size,1,True)
        self.output_conv = nn.Conv2d(hidden_dims[1], 1, (1, 1), padding='same')
    def forward(self, x):
        l1_o, _=self.cl1(x); l2_o,_=self.cl2(l1_o); return self.output_conv(l2_o[:,-1,:,:,:])

model = ConvLSTMNet(len(final_feature_vars))
if torch.cuda.device_count() > 1: model = nn.DataParallel(model)
model.to(device)
print("Model built and enabled for multi-GPU.")

# Step 5: Train and Visualize
# ------------------------------------------------------------------------------
def weighted_mse_loss(output, target, weight_threshold=0.8):
    mse = (output-target)**2; weights=torch.where(target>weight_threshold,2.0,1.0); return torch.mean(weights*mse)

def visualize_epoch_prediction(model, val_loader, device, scalers, epoch):
    model.eval()
    with torch.no_grad():
        X_val,y_val=next(iter(val_loader)); X_val=X_val.to(device); pred_val=model(X_val)[0].cpu().numpy().squeeze()
        true_val=y_val[0].cpu().numpy().squeeze(); ts=scalers[target_var]
        pred_orig=ts.inverse_transform(pred_val.reshape(1,-1)).reshape(pred_val.shape)
        true_orig=ts.inverse_transform(true_val.reshape(1,-1)).reshape(true_val.shape)
        fig, axes=plt.subplots(1,3,figsize=(20,5)); fig.suptitle(f'Epoch {epoch+1}-Sample Validation',fontsize=16)
        vmax=max(true_orig.max(),pred_orig.max())
        im1=axes[0].imshow(true_orig,cmap='viridis',vmin=0,vmax=vmax); axes[0].set_title("Truth"); plt.colorbar(im1,ax=axes[0],fraction=0.046,pad=0.04)
        im2=axes[1].imshow(pred_orig,cmap='viridis',vmin=0,vmax=vmax); axes[1].set_title("Prediction"); plt.colorbar(im2,ax=axes[1],fraction=0.046,pad=0.04)
        err=pred_orig-true_orig; vmax_err=np.abs(err).max()
        im3=axes[2].imshow(err,cmap='RdBu_r',vmin=-vmax_err,vmax=vmax_err); axes[2].set_title("Error"); plt.colorbar(im3,ax=axes[2],fraction=0.046,pad=0.04)
        plt.tight_layout(rect=[0,0.03,1,0.95]); plt.show()

def train_model(model, train_loader, val_loader, device, scalers, epochs=50, model_path='best_convlstm_model.pth'):
    optimizer=optim.Adam(model.parameters(),lr=1e-4); loss_fn=weighted_mse_loss
    best_val_loss=float('inf'); scaler=torch.cuda.amp.GradScaler()
    print("\n--- Step 5: Starting Model Training ---")
    for epoch in range(epochs):
        start_time=time.time(); model.train(); total_train_loss=0.0
        train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [Training]")
        for X, y in train_pbar:
            X, y = X.to(device), y.to(device)
            with torch.cuda.amp.autocast(): loss = loss_fn(model(X), y)
            optimizer.zero_grad(); scaler.scale(loss).backward(); scaler.step(optimizer); scaler.update()
            total_train_loss += loss.item()
            train_pbar.set_postfix({'loss': f'{loss.item():.6f}'})
        
        model.eval(); total_val_loss = 0.0
        with torch.no_grad():
            for X_v, y_v in val_loader:
                total_val_loss += loss_fn(model(X_v.to(device)), y_v.to(device)).item()
        
        avg_train_loss,avg_val_loss = total_train_loss/len(train_loader), total_val_loss/len(val_loader)
        print(f"Epoch {epoch+1} Summary | Avg Train Loss: {avg_train_loss:.6f} | Avg Val Loss: {avg_val_loss:.6f} | Time: {time.time()-start_time:.2f}s")
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            state_dict = model.module.state_dict() if isinstance(model, nn.DataParallel) else model.state_dict()
            torch.save(state_dict, model_path)
            print(f"  -> âœ… New best model saved (Val Loss: {best_val_loss:.6f})")
        
        if (epoch + 1) % 5 == 0:
            visualize_epoch_prediction(model, val_loader, device, scalers, epoch)
    print("\n--- Training finished! ---")

# --- Start the FULL Training ---
# I've set it to 50 epochs, but you can increase it.
train_model(model, train_loader, val_loader, device, scalers, epochs=50)
```

```python
# ==============================================================================
# FINAL and CORRECTED Cell for Evaluation (Architecture & All Fixes Included)
# ==============================================================================

# --- 1. Setup and Imports ---
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('Agg') # Use non-GUI backend
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tqdm.auto import tqdm
import warnings
import os

warnings.filterwarnings('ignore')
print("--- Step 1: Setup Complete ---")

# --- 2. Redefine Necessary Variables and Classes ---
# Configuration
data_path = r'D:\\babe_prediction\\cmems_mod_ibi_wav_anfc_0.027deg_PT1H-i_multi-vars_11.00W-8.53W_38.50N-40.97N_2022-12-01-2025-01-01-001.nc'
target_var = 'VHM0'
initial_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR', 'VTPK', 'VSDX', 'VSDY']
final_feature_vars = ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR_sin', 'VMDR_cos', 'VTPK', 'VSDX', 'VSDY']
all_vars_to_process = [target_var] + initial_feature_vars
lookback_hours = 24
batch_size = 8
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# OPTIMIZED: Re-fit Scalers in Memory-Efficient Chunks
print("--- Step 2: Re-fitting scalers memory-efficiently... ---")
ds = xr.open_dataset(data_path)
scalers = {}
chunk_size = 1000

for var in all_vars_to_process:
    if var == 'VMDR': continue
    scaler = MinMaxScaler()
    for i in tqdm(range(0, ds.dims['time'], chunk_size), desc=f"Fitting scaler for {var}"):
        chunk = ds[var].isel(time=slice(i, i + chunk_size)).fillna(0).values.reshape(-1, 1)
        scaler.partial_fit(chunk)
    scalers[var] = scaler
print("Scalers are ready.")

# Memory-Efficient Lazy Dataset Class
class WaveDataset_Lazy(Dataset):
    def __init__(self, data_path, feature_vars, target_var, indices, scalers, lookback=24):
        self.data_path, self.feature_vars, self.target_var, self.indices, self.lookback, self.scalers = data_path, feature_vars, target_var, indices, lookback, scalers
        self.all_vars_on_disk = [target_var] + ['VHM0_SW1', 'VHM0_SW2', 'VHM0_WW', 'VMDR', 'VTPK', 'VSDX', 'VSDY']
    def __len__(self): return len(self.indices)
    def __getitem__(self, idx):
        seq_start_idx = self.indices[idx]; target_idx = seq_start_idx + self.lookback
        with xr.open_dataset(self.data_path) as ds_slice:
            data_slice = ds_slice[self.all_vars_on_disk].isel(time=slice(seq_start_idx, target_idx + 1))
        
        data_slice = data_slice.fillna(0)
        radians = np.deg2rad(data_slice['VMDR'].values)
        data_slice['VMDR_sin'] = (('time', 'latitude', 'longitude'), np.sin(radians).astype(np.float32))
        data_slice['VMDR_cos'] = (('time', 'latitude', 'longitude'), np.cos(radians).astype(np.float32))

        for var, scaler in self.scalers.items():
            var_data = data_slice[var].values; original_shape = var_data.shape
            var_data_flattened = var_data.reshape(-1, 1)
            var_scaled_flattened = scaler.transform(var_data_flattened)
            var_scaled_reshaped = var_scaled_flattened.reshape(original_shape)
            data_slice[var] = (data_slice[var].dims, var_scaled_reshaped)

        features_list = [data_slice[var].isel(time=slice(0, self.lookback)).values for var in self.feature_vars]
        X = np.stack(features_list, axis=0).astype(np.float32)
        y = data_slice[self.target_var].isel(time=self.lookback).values.astype(np.float32)
        X, y = X.transpose(1, 0, 2, 3), np.expand_dims(y, axis=0)
        return torch.from_numpy(X), torch.from_numpy(y)

# --- CORRECTED: Full Custom Model Architecture ---
class ConvLSTMCell(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, bias):
        super(ConvLSTMCell, self).__init__(); self.input_dim,self.hidden_dim,self.kernel_size,self.padding,self.bias = input_dim,hidden_dim,kernel_size,kernel_size[0]//2,bias
        self.conv = nn.Conv2d(self.input_dim+self.hidden_dim, 4*self.hidden_dim, self.kernel_size, padding=self.padding, bias=self.bias)
    def forward(self, x, h_c):
        h,c=h_c; cc=self.conv(torch.cat([x,h],dim=1)); i,f,o,g=torch.split(cc,self.hidden_dim,dim=1)
        i,f,o,g=torch.sigmoid(i),torch.sigmoid(f),torch.sigmoid(o),torch.tanh(g); c_n=f*c+i*g
        return o*torch.tanh(c_n), c_n
    def init_hidden(self, b, i_s, d): return (torch.zeros(b,self.hidden_dim,*i_s,device=d), torch.zeros(b,self.hidden_dim,*i_s,device=d))

class ConvLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, kernel_size, num_layers, batch_first=False, bias=True):
        super(ConvLSTM, self).__init__(); self.batch_first=batch_first; self.num_layers=num_layers
        h_d=[hidden_dim]*num_layers if isinstance(hidden_dim,int) else hidden_dim
        cl=[ConvLSTMCell(input_dim if i==0 else h_d[i-1],h_d[i],kernel_size,bias) for i in range(num_layers)]
        self.cell_list=nn.ModuleList(cl)
    def forward(self, x, h_c=None):
        b,_,_,h,w=x.size(); h_c=self._init_hidden(b,(h,w),x.device) if h_c is None else h_c
        out_l,last_s=[],[]; s_l=x.size(1); cur_in=x
        for l_idx in range(self.num_layers):
            h,c=h_c[l_idx]; out_i=[]
            for t in range(s_l):
                h,c=self.cell_list[l_idx](cur_in[:,t,:,:,:], [h,c]); out_i.append(h)
            l_out=torch.stack(out_i,dim=1); cur_in=l_out; out_l.append(l_out); last_s.append([h,c])
        return out_l[-1], last_s[-1]
    def _init_hidden(self, b, i_s, d): return [c.init_hidden(b,i_s,d) for c in self.cell_list]

class ConvLSTMNet(nn.Module):
    def __init__(self, input_dim, hidden_dims=[64, 32], kernel_size=(3, 3)):
        super(ConvLSTMNet, self).__init__()
        # THIS IS THE CORRECTED PART: It now calls our custom ConvLSTM class
        self.cl1 = ConvLSTM(input_dim, hidden_dims[0], kernel_size, 1, batch_first=True)
        self.cl2 = ConvLSTM(hidden_dims[0], hidden_dims[1], kernel_size, 1, batch_first=True)
        self.output_conv = nn.Conv2d(hidden_dims[1], 1, kernel_size=(1, 1), padding='same')
    def forward(self, x):
        l1_o, _ = self.cl1(x)
        l2_o, _ = self.cl2(l1_o)
        return self.output_conv(l2_o[:, -1, :, :, :])

# --- 3. Load the Trained Model ---
print("--- Step 3: Loading the best trained model ---")
final_model = ConvLSTMNet(len(final_feature_vars))
model_path = 'best_convlstm_model.pth'
try:
    final_model.load_state_dict(torch.load(model_path, map_location=device))
    if torch.cuda.device_count() > 1:
        final_model = nn.DataParallel(final_model)
    final_model.to(device)
    final_model.eval()
    print(f"--- âœ… Model successfully loaded from {model_path} ---")
except Exception as e:
    raise RuntimeError(f"Could not load the model. Ensure 'best_convlstm_model.pth' exists and matches the architecture. Error: {e}")

# --- 4. Create the Test DataLoader ---
print("--- Step 4: Preparing Test Data ---")
n_sequences = ds.dims['time'] - lookback_hours
all_indices = np.arange(n_sequences)
train_end, val_end = int(n_sequences * 0.7), int(n_sequences * 0.85)
test_indices = all_indices[val_end:]
test_dataset = WaveDataset_Lazy(data_path, final_feature_vars, target_var, test_indices, scalers, lookback_hours)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=0)

# --- 5. Run Evaluation and Visualize ---
print("\n--- Step 5: Starting Final Evaluation ---")
all_y_true, all_y_pred = [], []
with torch.no_grad():
    for X_batch, y_batch in tqdm(test_loader, desc="Evaluating on Test Set"):
        y_pred = final_model(X_batch.to(device))
        all_y_true.append(y_batch.cpu().numpy())
        all_y_pred.append(y_pred.cpu().numpy())

y_test_scaled = np.concatenate(all_y_true).squeeze(axis=1)
y_pred_scaled = np.concatenate(all_y_pred).squeeze(axis=1)

# Inverse transform and calculate metrics
target_scaler = scalers[target_var]
y_test_orig = target_scaler.inverse_transform(y_test_scaled.reshape(len(y_test_scaled), -1)).reshape(y_test_scaled.shape)
y_pred_orig = target_scaler.inverse_transform(y_pred_scaled.reshape(len(y_pred_scaled), -1)).reshape(y_pred_scaled.shape)

test_rmse = np.sqrt(mean_squared_error(y_test_orig.flatten(), y_pred_orig.flatten()))
test_mae = mean_absolute_error(y_test_orig.flatten(), y_pred_orig.flatten())

print(f"\n--- Test Set Performance ---")
print(f"  -> Root Mean Squared Error (RMSE): {test_rmse:.4f} m")
print(f"  -> Mean Absolute Error (MAE):    {test_mae:.4f} m")

# --- Visualizations ---
print("\n--- Generating visualizations... ---")
# Spatial Map Comparison
time_idx_to_plot = 150
true_map, predicted_map = y_test_orig[time_idx_to_plot], y_pred_orig[time_idx_to_plot]
error_map = predicted_map - true_map
fig, axes = plt.subplots(1, 3, figsize=(22, 6)); fig.suptitle('Spatial Map Comparison (Test Set Sample)', fontsize=16)
vmax = max(true_map.max(), predicted_map.max())
im1=axes[0].imshow(true_map, cmap='viridis', vmin=0, vmax=vmax); axes[0].set_title("Ground Truth"); plt.colorbar(im1, ax=axes[0], label='Wave Height (m)', fraction=0.046, pad=0.04)
im2=axes[1].imshow(predicted_map, cmap='viridis', vmin=0, vmax=vmax); axes[1].set_title("Model Prediction"); plt.colorbar(im2, ax=axes[1], label='Wave Height (m)', fraction=0.046, pad=0.04)
vmax_err = np.abs(error_map).max()
im3=axes[2].imshow(error_map, cmap='RdBu_r', vmin=-vmax_err, vmax=vmax_err); axes[2].set_title("Prediction Error"); plt.colorbar(im3, ax=axes[2], label='Error (m)', fraction=0.046, pad=0.04)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

# Time-Series Plot
central_lat_idx, central_lon_idx = true_map.shape[0]//2, true_map.shape[1]//2
true_series, pred_series = y_test_orig[:, central_lat_idx, central_lon_idx], y_pred_orig[:, central_lat_idx, central_lon_idx]
plt.figure(figsize=(18, 7))
plt.plot(true_series, label='True VHM0 (Ground Truth)', color='blue', alpha=0.9, linewidth=2)
plt.plot(pred_series, label='Predicted VHM0', color='red', linestyle='--', alpha=0.8)
plt.title('Time-Series Comparison at a Central Grid Point (Test Set)')
plt.xlabel('Time Step (Hour) in Test Set'); plt.ylabel('Significant Wave Height (m)')
plt.legend(); plt.grid(True, linestyle='--', alpha=0.6); plt.show()
```
